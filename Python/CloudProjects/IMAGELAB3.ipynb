```python
#Author: Hubert Dziurla
#The function should contain: direction
#Import cv2 scripts with patches for imshow 
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
#Step 1 Read an image
image = cv2.imread("LineFollowerTest.bmp");
#Step 2 Convert image RGB color to gray to better recognition.
imag2 = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY);
#Step 2 Use Canny operator
imageEdge = cv2.Canny(imag2,100,100);
#Step 3 Use Hough Transform
lines = cv2.HoughLines(imageEdge,1,np.pi/90,200)
cv2_imshow(imageEdge)

   
```
